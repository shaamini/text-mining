```python
import os
import pandas as pd

masterdf = pd.DataFrame()

for file in os.listdir(os.getcwd()):
    if file.endswith('.csv'):
        masterdf = masterdf.append(pd.read_csv(file))

masterdf.to_csv('tfidf-descending-india-19000-38000.csv', index=False)
```
