```python
import nltk
import random
from nltk.corpus import CategorizedPlaintextCorpusReader
```

```python
#Create the list of documents with the categories. 
#Below code combines all the files and stores randomly.
documents = []
for i in range(len(x_train1)):
    each_x = x_train1.iloc[i]
    x_words = ast.literal_eval(each_x["words_processed_all"])
    
    each_y = y_train1.iloc[i]
    y_category = each_y["text_type"]
    
    word_category = (x_words, y_category)
    documents.append(word_category)
random.shuffle(documents)
```

```python
#The below code is for the feature set definition. We are using only top 2000 words as our features 
all_words = []
for each in x_train1["words_processed_all"]:
    words = ast.literal_eval(each)
    for each2 in words:
        all_words.append(each2)
        
all_words = nltk.FreqDist(w for w in all_words)

word_features = list(all_words)[100:2000]
```

```python
word_features
```

```python
#The below code is the function, feature extractor
def document_features(document): 
    document_words = set(document)
    features = {}
    for word in word_features:
        features['contains({})'.format(word)] = (word in document_words)
    return features
#print(document_features(sgnews.words('crime/Crime_9067.txt')))

#We can reuse the previous classifier and test. But we didn't train it with the feature set. So 
#We now divide the train set into test and train set. We can use different combinations. Call classifier
featuresets = [(document_features(d), c) for (d,c) in documents]
train_set, test_set = featuresets[100:], featuresets[:100]
classifier = nltk.NaiveBayesClassifier.train(train_set)

#Print the accuray and the most informative features
print(nltk.classify.accuracy(classifier, test_set))
classifier.show_most_informative_features(5)

#What this tells you is the ratio of occurences in Sports to Crime, or visa versa, 
#for every word. So here, we can see that the term "police" appears 12.3 more times 
#as often in Crime articles as it does in Sports articles.

print(classifier.labels())
```

```python
#Precision and Recall

test_result = []
gold_result = []

for i in range(len(test_set)):
    test_result.append(classifier.classify(test_set[i][0]))
    gold_result.append(test_set[i][1])
    
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix   

print('\nClasification report:\n', classification_report(gold_result, test_result))
print('\nConfussion matrix:\n',confusion_matrix(gold_result, test_result)  )
```

```python
def simple_logistic_classify(X_tr, y_tr, X_test, y_test, description, _C=1.0):
    model = LogisticRegression(C=_C).fit(X_tr, y_tr)
    score = model.score(X_test, y_test)
    print('Test Score with', description, 'features', score)
    return model
```

```python
model_bow = simple_logistic_classify(X_tr_bow, y_tr, X_te_bow, y_te, 'bow')
model_tfidf = simple_logistic_classify(X_tr_tfidf, y_tr, X_te_tfidf, y_te, 'tf-idf')
```
