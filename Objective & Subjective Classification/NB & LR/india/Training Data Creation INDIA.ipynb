```python
df = df.iloc[:1000]
df
```

```python
#Trying to use subjectivity in helping me manually label the documents
from textblob import TextBlob
df['subjectivity']=df['translated'].apply(lambda x:TextBlob(x).sentiment.subjectivity)
```

```python
text_type = []
for each in df["subjectivity"]:
    if each >= 0.5:
        text_type.append("subjective")
    else:
        text_type.append("objective")
df["text_type"] = text_type
```

```python
objective = df[df["text_type"] == "objective"]
subjective = df[df["text_type"] == "subjective"]
```
