# Read Data (Labelled & Whole)

```python
import pandas as pd
import numpy as np
```

```python
labelled_df = pd.read_csv("training_data_india.csv")
labelled_df.drop(['Unnamed: 0', 'Unnamed: 0.1'], axis=1, inplace=True)
labelled_df
```

```python
whole_df = pd.read_csv("../../india-v1-preprocessed-overall.csv")
whole_df.drop(['Unnamed: 0'], axis=1, inplace=True)
whole_df
```

# Naive Bayes

## Train Model

```python
X = labelled_df.translated
y = labelled_df.text_type
```

```python
#split into training and testing datasets
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1, test_size = 0.2)
print(X_train.shape)
print(X_test.shape)
```

```python
# instantiate the vectorizer
from sklearn.feature_extraction.text import CountVectorizer

v = CountVectorizer()
```

```python
#Learn vocabulary from the entire dataset rather than just the labelled df
X_whole_data = v.fit_transform(whole_df.translated)
X_whole_data
```

```python
# transform testing data (using fitted vocabulary) into a document-term matrix
X_test_data = v.transform(X_test)
X_train_data = v.transform(X_train)
```

```python
X_test_data
```

```python
X_train_data
```

```python
# train a Naive Bayes model using X_train_data and Xf_train_data
from sklearn.naive_bayes import MultinomialNB

naive_bayes = MultinomialNB()

naive_bayes.fit(X_train_data, y_train)
```

```python
# make class predictions  
y_pred_class = naive_bayes.predict(X_test_data)
```

```python
# calculate accuracy of class predictions from the one-time cross validation
# compute the accuracy scores
from sklearn import metrics
from sklearn.metrics import f1_score
NB = metrics.accuracy_score(y_test, y_pred_class)
NB_fscore = f1_score(y_test, y_pred_class, average='weighted')
print("Accuracy: " + str(NB*100))
print("F1 Score: " + str(NB_fscore*100))
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred_class)
print(cm)
```

```python
y_pred_class
```

## Analysing misclassified sentences

```python
indexes = list(X_test.index)
test_df = labelled_df.iloc[indexes]
test_df["predicted"] = list(y_pred_class)
test_df
```

```python
misclassified = test_df[test_df["text_type"]!=test_df["predicted"]]
```

```python
for i in range(len(misclassified)):
    each = misclassified.iloc[i]
    wrong = each["predicted"]
    correct = each["text_type"]
    sentence = each["translated"]
    print(sentence)
    print(correct)
    print(wrong)
    print(" ")
```

## Save the model

```python
import pickle
save_model = open("india_nb.pickle","wb") #binary write
pickle.dump(naive_bayes, save_model)
save_model.close()
```
