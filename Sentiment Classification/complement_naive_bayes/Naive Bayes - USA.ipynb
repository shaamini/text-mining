```python
# Helper function to remove unwanted patterns
def remove_pattern(input_txt, pattern):
    r = re.findall(pattern, input_txt)
    for i in r:
        input_txt = re.sub(i, '', input_txt)
    return input_txt

# Remove Twitter handles from the data 
tweets['new_text_after_translation'] = np.vectorize(remove_pattern)(tweets['new_text_after_translation'], "@[\w]*")

# Remove punctuations, numbers, and special characters
tweets['new_text_after_translation'] = tweets['new_text_after_translation'].str.replace("[^a-zA-Z]", " ")
```
