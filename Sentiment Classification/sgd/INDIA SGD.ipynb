### Read data (Labelled)

```python
labelled_df = pd.read_csv("../../1k_sample_labelled_india.csv")
# labelled_df.drop(['Unnamed: 0'], axis=1, inplace=True)
labelled_df
```

```python
def sentiment2target(sentiment):
    return {
        'negative': 0,
        'neutral': 1,
        'positive' : 2
    }[sentiment]
targets = labelled_df.sentiment_vader.apply(sentiment2target)
```

```python
labelled_df['sentiment_vader'] =targets
```

```python
labelled_df
```

## SGD

```python
from sklearn.feature_extraction.text import CountVectorizer
from nltk.tokenize import RegexpTokenizer
token = RegexpTokenizer(r'[a-zA-Z0-9]+')
cv = CountVectorizer(stop_words='english',ngram_range = (1,1),tokenizer = token.tokenize)
X_whole_data = cv.fit_transform(india_opinion['new_text'].values.astype('U'))
```

```python
from sklearn.model_selection import train_test_split
# X=text_counts
X = cv.transform(labelled_df.new_text)
y=labelled_df.sentiment_vader
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30,random_state=19)
```

```python
print(X_train.shape)
print(X_test.shape)
```

```python
import scipy.sparse as sp
train_count = sp.csr_matrix(X_train)
```

```python
X_train = sp.hstack([train_count])
X_train
```

```python
test_count = sp.csr_matrix(X_test)
X_test = sp.hstack([test_count])
```

```python
from sklearn.linear_model import SGDClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score
from sklearn.metrics import confusion_matrix
n_classes = 3

clf = SGDClassifier(alpha=1e-3, random_state=0)
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)
```

```python
from sklearn.metrics import *
#Checking performance our model with classification report
print(classification_report(y_test, y_pred))
```

```python
from sklearn.metrics import *
from sklearn.metrics import *
# from sklearn.metrics import confusion_matrix
n_classes = 3

cm = confusion_matrix(y_test, y_pred, labels=range(n_classes))

print(f'Number of samples to classify: {len(X_test.toarray())}\n')

print("Accuracy score ={:.2f}%".format(accuracy_score(y_test, y_pred)*100))
print("F1 score ={:.2f}%".format(f1_score(y_test, y_pred, average="macro")*100))
print("Precision score ={:.2f}%".format(precision_score(y_test, y_pred, average="macro")*100))
print("Recall score ={:.2f}%".format(recall_score(y_test, y_pred, average="macro")*100))  
```

```python
import pickle
with open('india_sgd.model', "wb") as f:
 pickle.dump(clf, f)
 f.close()
```
