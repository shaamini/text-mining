## 6 topics

```python
# TRY DIFFERENT NUM TOPICS
# keep beta = 0.09, try different beta if results arent great
# iterations = 600 seems decent

X, vocabulary, vocab_dict = btm.get_words_freqs(texts, stop_words=stop_words)
docs_vec = btm.get_vectorized_docs(texts, vocabulary)
biterms = btm.get_biterms(docs_vec)

# Initializing and running model
model = btm.BTM(
    X, vocabulary, seed=12321, T=6, M=20, alpha=50/6, beta=0.09)
model.fit(biterms, iterations=600)
```

```python
entropy = btm.entropy(model.matrix_topics_words_)
entropy
```

```python
p_zd = model.transform(docs_vec)
coherence = model.coherence_
coherence
```

```python
import tmplot as tmp

# Run the interactive report interface
tmp.report(model=model, docs=texts)
```

```python
stable_topics = [0, 1, 2,3,4,5]
top_words = btm.get_top_topic_words(
    model,
    words_num=12,
    topics_idx=stable_topics)

top_words
```

## Checking 10 topics

```python
matrix10 = pd.DataFrame(data=p_zd10,columns=['Topic0','Topic1','Topic2','Topic3','Topic4','Topic5','Topic6','Topic7','Topic8','Topic9'])
```

```python
matrix10
```

```python
p_zd10
```

```python
dominant_topic = []
for each in p_zd10:
    topic = np.argmax(each)
    dominant_topic.append(topic)
```

```python
matrix10['dominant'] = dominant_topic
```

```python
matrix10
```

```python
sum_up_to_one = []

for each in p_zd10:
    if sum(each) == 0:
        result = "Unclassified"
    else:
        result = "Classified"
    sum_up_to_one.append(result)
```

```python
matrix10['sum_up'] = sum_up_to_one
```

```python
matrix10[matrix10['sum_up']=='Unclassified']
```

```python
matrix10.to_csv('check_matrix10.csv')
```
