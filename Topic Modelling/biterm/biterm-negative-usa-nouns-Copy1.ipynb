```python
import bitermplus as btm
import numpy as np
import pandas as pd
```

```python
# Importing data
df = pd.read_csv(
    '../pos_neg_usa.csv')
```

```python
df
```

```python
df.drop(['Unnamed: 0'], axis=1, inplace=True)
```

```python
df = df[df['sentiment']==0]
```

```python
df.head()
```

```python
df.reset_index(inplace=True)
```

```python
df.drop(['index'], axis=1, inplace=True)
```

```python
df
```

```python
texts = df['words_processed_noun']
```

```python
texts
```

```python
new_text_noun = []

for row in texts:
    new_row = row.strip("][").replace("'","").split(", ")
    new_row = " ".join(new_row)
    new_text_noun.append(new_row)
```

```python
new_text_noun
```

```python
df['new_text_noun'] = new_text_noun
```

```python
texts = df['new_text_noun']
```

```python
texts
```

```python
# Vectorizing documents, obtaining full vocabulary and biterms
# Internally, btm.get_words_freqs uses CountVectorizer from sklearn
# You can pass any of its arguments to btm.get_words_freqs
# For example, you can remove stop words:
stop_words = ['vaccine','people','couid','lot','time','today','vaccination','covidvaccine', 'day','go','one',
             'get', 'not', 'like', 'make', 'take', 'hai', 'ki', 'se', 'ko', 'ka', 'ke', 'hi', 'ho', 'vaccinated',
             'vaccinate', 'vaccinated', 'amp', 'two', 'even', 'thing', 'covishield', 'yesterday', 'pfizer',
             'nineteen', 'crore', 'dose', 'dos', 'booster', 'world', 'health', 'new', 'also', 'app', 'done',
             'second', 'still', 'need', 'three', 'please', 'getting', 'make', 'five', 'six', 'seven', 'eight', 
             'nine', 'ten', 'taken', 'est', 'say', 'given', 'fully', 'made', 'number', 'person', 'home',
             'fact', 'hand', 'week', 'month', 'year', 'life', 'govt', 'kid', 'parent', 'child', 'doctor', 'nurse',
             'patient', 'friend', 'mother', 'teacher', 'government', 'state', 'country', 'hospital', 'student',
             'photo', 'family', 'use', 'cow', 'test', 'case', 'sir', 'pal', 'player', 'video', 'cat', 'way',
             'place', 'staff', 'mom', 'man', 'video', 'love', 'call', 'smile', 'ahmedabad', 'thank', 'hour',
             'other', 'part', 'shit', 'lie', 'word', 'point', 'side', 'group', 'fan', 'room', 'dog', 'play',
             'food', 'wife', 'net', 'face', 'car', 'ass', 'trust', 'eye', 'bar', 'sport', 'age', 'era']
```

## 4 topics

```python
# TRY DIFFERENT NUM TOPICS
# keep beta = 0.09, try different beta if results arent great
# iterations = 600 seems decent

X, vocabulary, vocab_dict = btm.get_words_freqs(texts, stop_words=stop_words)
docs_vec = btm.get_vectorized_docs(texts, vocabulary)
biterms = btm.get_biterms(docs_vec)

# Initializing and running model
model = btm.BTM(
    X, vocabulary, seed=12321, T=4, M=20, alpha=50/4, beta=0.09)
model.fit(biterms, iterations=600)
```

```python
entropy = btm.entropy(model.matrix_topics_words_)
```

```python
entropy
```

## Calculate documents vs topics probability matrix

```python
p_zd = model.transform(docs_vec)
```

```python
coherence = model.coherence_
```

```python
coherence
```

```python
import tmplot as tmp

# Run the interactive report interface
tmp.report(model=model, docs=texts)
```
