```python
import pickle
model_saved = open("india_biterm.model", "rb") #binary read
biterm = pickle.load(model_saved)
model_saved.close()
```

```python
new_texts = ['death rate harm harm']
```

```python
new_docs_vec = btm.get_vectorized_docs(new_texts, vocabulary)
p_zd = biterm.transform(new_docs_vec)
```

```python
if max(p_zd[0]) > 0.3:
    print(np.argmax(p_zd))
else:
    print("unclassified")
```
